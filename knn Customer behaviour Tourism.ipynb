```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load the dataset from CSV
file_path = 'Customer behaviour Tourism.csv'  # Adjust the file path accordingly
data = pd.read_csv(file_path)
```

```python
data.head()

```

```python
# Data Preprocessing
label_encoder = LabelEncoder()
data['preferred_device'] = label_encoder.fit_transform(data['preferred_device'])
data['preferred_location_type'] = label_encoder.fit_transform(data['preferred_location_type'])
data['following_company_page'] = label_encoder.fit_transform(data['following_company_page'])
data['working_flag'] = label_encoder.fit_transform(data['working_flag'])
data['Taken_product'] = label_encoder.fit_transform(data['Taken_product'])

# Replace non-numeric values with NaN
data = data.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values
data = data.dropna()
```

```python
# Split the data into features (X) and target variable (y)
X = data.drop(['UserID', 'Taken_product'], axis=1)
y = data['Taken_product']
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Try different values of k
for k in range(1, 11):  # You can adjust the range as needed
    # Build a k-NN Model
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train, y_train)

    # Model Evaluation
    y_pred = knn_model.predict(X_test)

    # Calculate and display relevant metrics for each k
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    # Display results for each k
    print(f'For k = {k}:')
    print(f'Accuracy: {accuracy:.2f}')
    print(f'Confusion Matrix:\n{conf_matrix}')
    print(f'Classification Report:\n{class_report}')
    print('-' * 50)
```
